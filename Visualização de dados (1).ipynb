```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.set_palette("Set2")
```

```python
arq = pd.read_csv('vgsales.csv')
arq
```

### Gráfico de dispersão para calcular a correlação e tendência de duas variáveis

```python
sns.scatterplot(data=arq, x="Year", y="Other_Sales")
```

```python
sns.heatmap(arq.corr(), annot=True)
```

### Histograma

```python
sns.displot(arq, x="Year", bins=10)
```

### Gráfico de linha

```python
sns.lineplot(data=arq, x="Year", y="JP_Sales")
```
